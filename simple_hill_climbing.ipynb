# simple hill climbing

```python
import random  # Random numbers generate karne ke liye library

# Objective function (maximize karna hai)
def objective_function(x):
    return -(x - 3) ** 2 + 9  # Peak at x=3, max value = 9

# Hill climbing algorithm
def hill_climbing():
    current_solution = random.uniform(-10, 10)  # Random initial solution
    current_value = objective_function(current_solution)  # Evaluate initial solution

    step_size = 0.1 # Step size thoda bada rakha taake jaldi converge ho

    for i in range(10):  # Sirf 10 iterations chalenge
        new_solution = current_solution + random.choice([-step_size, step_size])  # Move left/right
        new_value = objective_function(new_solution)  # Evaluate new solution

        if new_value > current_value:  # Agar better hai to accept karo
            current_solution = new_solution
            current_value = new_value

        print(f"Iteration {i}: x = {current_solution:.2f}, value = {current_value:.2f}")  # Progress print

    return current_solution, current_value  # Best solution return

# Run algorithm
best_x, best_value = hill_climbing()
print(f"Best solution found: x = {best_x:.2f}, value = {best_value:.2f}")  # Final result

```
